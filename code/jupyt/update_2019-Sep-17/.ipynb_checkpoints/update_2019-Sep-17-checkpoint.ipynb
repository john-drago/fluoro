```python
def plot_label_distributions(labels_data, plot_title, plot_save_name):
    
    fig1, axs = plt.subplots(3,2, figsize=(15,15))
    plt.subplots_adjust(wspace=0.4, hspace=0.4)

    plt.rc('text', usetex=True)

    axs[0,0].hist(labels_data[:,0],50, density = True, edgecolor='black')
    axs[0,0].set_title(r'\textbf{Rotation: $\theta$}', fontweight = 'bold')
    axs[0,0].grid(False)
    axs[0,0].set_ylabel('PDF')
    axs[0,0].set_xlabel('Angles (radians)')
    # axs[0,0].text(-1.55,.7, 
    #               r'$\mu = $' + str(np.around(labels_mean[0],3))+'\n'+ r'$\sigma = $' + str(np.around(labels_std[0],3)))

    axs[1,0].hist(labels_data[:,1],50, density = True, edgecolor='black')
    axs[1,0].set_title(r'\textbf{Rotation: $\phi$}', fontweight = 'bold')
    axs[1,0].grid(False)
    axs[1,0].set_ylabel('PDF')
    axs[1,0].set_xlabel('Angles (radians)')
    # axs[1,0].text(-1.55,1.2, 
    #               r'$\mu = $' + str(np.around(labels_mean[1],3))+'\n'+ r'$\sigma = $' + str(np.around(labels_std[1],3)))

    axs[2,0].hist(labels_data[:,2],50, density = True, edgecolor='black')
    axs[2,0].set_title(r'\textbf{Rotation: $\psi$}', fontweight = 'bold')
    axs[2,0].grid(False)
    axs[2,0].set_ylabel('PDF')
    axs[2,0].set_xlabel('Angles (radians)')
    # axs[2,0].text(-1.55, 0.8, 
    #               r'$\mu = $' + str(np.around(labels_mean[2],3))+'\n'+ r'$\sigma = $' + str(np.around(labels_std[2],3)))

    axs[0,1].hist(labels_data[:,3],50, color='red', density = True, edgecolor='black')
    axs[0,1].set_title(r'\textbf{Translation: $x$}', fontweight = 'bold')
    axs[0,1].grid(False)
    axs[0,1].set_ylabel('PDF')
    axs[0,1].set_xlabel('Distance (mm)')
    # axs[0,1].text(-140, 0.0075, 
    #               r'$\mu = $' + str(np.around(labels_mean[3],3))+'\n'+ r'$\sigma = $' + str(np.around(labels_std[3],3)))

    axs[1,1].hist(labels_data[:,4],50, color='red', density = True, edgecolor='black')
    axs[1,1].set_title(r'\textbf{Translation: $y$}', fontweight = 'bold')
    axs[1,1].grid(False)
    axs[1,1].set_ylabel('PDF')
    axs[1,1].set_xlabel('Distance (mm)')
    # axs[1,1].text(-120, 0.012, 
    #               r'$\mu = $' + str(np.around(labels_mean[4],3))+'\n'+ r'$\sigma = $' + str(np.around(labels_std[4],3)))

    axs[2,1].hist(labels_data[:,5],50, color='red', density = True, edgecolor='black')
    axs[2,1].set_title(r'\textbf{Translation: $z$}', fontweight = 'bold')
    axs[2,1].grid(False)
    axs[2,1].set_ylabel('PDF')
    axs[2,1].set_xlabel('Distance (mm)')
    # axs[2,1].text(0, 0.006, 
    #               r'$\mu = $' + str(np.around(labels_mean[5],3))+'\n'+ r'$\sigma = $' + str(np.around(labels_std[5],3)))

    fig1.suptitle(r'\textbf{' + plot_title + '}',fontsize=16, y = 0.93, fontweight= 'bold')
    fig1.show()
    fig1.savefig(os.path.join(image_dir, plot_save_name),dpi=300)
    
    return None



def plot_loss_curve(train_loss, val_loss, num_epochs, plot_title, plot_save_name):
    
    fig2, axs = plt.subplots(1,1, figsize=(15,6))

    plt.rc('text', usetex=True)

    axs.plot(np.arange(1,num_epochs+1),train_loss,'b-',label='Train')
    axs.plot(np.arange(1,num_epochs+1), val_loss, 'r-',label='Val')
    axs.set_xlabel('Epoch')
    axs.set_ylabel('Normalized MSE')
    axs.legend(loc="upper right", fontsize=14) 

    # axs[1].plot(np.arange(1,31),train_loss,'b-',label='Train')
    # axs[1].plot(np.arange(1,31), val_loss, 'r-',label='Val')
    # axs[1].set_xlabel('Epoch')
    # axs[1].set_ylabel('Normalized MSE')
    # axs[1].legend(loc="upper right", fontsize=14) 
    # axs[1].set_ylim(2.6,4)

    fig2.show()
    fig2.suptitle(r'\textbf{' + plot_title + '}',fontsize=16,y=0.95)

    fig2.savefig(os.path.join(image_dir, plot_save_name), dpi=300)
    
    return None



def plot_prediction_v_actual(prediction_array, target_array, plot_title, plot_save_name, width = 0.3):
    
    fig2, axs = plt.subplots(3,2, figsize=(15,15))
    plt.subplots_adjust(wspace=0.4, hspace=0.4)

    plt.rc('text', usetex=True)

    labels = np.arange(1,prediction_array.shape[0]+1)

    axs[0,0].bar((labels)-width/2, prediction_array[:,0], width, label='Prediction')
    axs[0,0].bar((labels)+width/2, target_array[:,0], width, label='Actual')
    axs[0,0].set_title(r'\textbf{Prediction v. Actual:} $\theta$', fontweight = 'bold')
    axs[0,0].legend(loc="lower right", fontsize=14) 
    axs[0,0].set_xticks(labels)
    axs[0,0].set_xlabel('Instance')
    axs[0,0].set_ylabel('Angle (radians)')

    axs[1,0].bar((labels)-width/2, prediction_array[:,1], width, label='Prediction')
    axs[1,0].bar((labels)+width/2, target_array[:,1], width, label='Actual')
    axs[1,0].set_title(r'\textbf{Prediction v. Actual:} $\phi$', fontweight = 'bold') 
    axs[1,0].set_xticks(labels)
    axs[1,0].set_xlabel('Instance')
    axs[1,0].set_ylim(-1.52,0.1)
    axs[1,0].set_ylabel('Angle (radians)')

    axs[2,0].bar((labels)-width/2, prediction_array[:,2], width, label='Prediction')
    axs[2,0].bar((labels)+width/2, target_array[:,2], width, label='Actual')
    axs[2,0].set_title(r'\textbf{Prediction v. Actual:} $\psi$', fontweight = 'bold') 
    axs[2,0].set_xticks(labels)
    axs[2,0].set_xlabel('Instance')
    axs[2,0].set_ylabel('Angle (radians)')

    axs[0,1].bar((labels)-width/2, prediction_array[:,3], width, label='Prediction')
    axs[0,1].bar((labels)+width/2, target_array[:,3], width, label='Actual')
    axs[0,1].set_title(r'\textbf{Prediction v. Actual:} $x$', fontweight = 'bold')
    axs[0,1].set_xticks(labels)
    axs[0,1].set_xlabel('Instance')
    axs[0,1].set_ylabel('Distance (mm)')

    axs[1,1].bar((labels)-width/2, prediction_array[:,4], width, label='Prediction')
    axs[1,1].bar((labels)+width/2, target_array[:,4], width, label='Actual')
    axs[1,1].set_title(r'\textbf{Prediction v. Actual:} $y$', fontweight = 'bold')
    axs[1,1].set_xticks(labels)
    axs[1,1].set_xlabel('Instance')
    axs[1,1].set_ylabel('Distance (mm)')

    axs[2,1].bar((labels)-width/2, prediction_array[:,5], width, label='Prediction')
    axs[2,1].bar((labels)+width/2, target_array[:,5], width, label='Actual')
    axs[2,1].set_title(r'\textbf{Prediction v. Actual:} $z$', fontweight = 'bold')
    axs[2,1].set_xticks(labels)
    axs[2,1].set_xlabel('Instance')
    axs[2,1].set_ylabel('Distance (mm)')

    fig2.suptitle(r'\textbf{' + plot_title + '}',fontsize=16, y = 0.93, fontweight= 'bold')
    fig2.show()
    fig2.savefig(os.path.join(image_dir, plot_save_name), dpi=300)    

    return None




def plot_distribution_of_prediction_error(prediction_array, target_array, plot_title, plot_save_name):
    
    loss_dist_mat = target_array - prediction_array 
    loss_dist_mat_mean = np.mean(loss_dist_mat,axis=0)
    loss_dist_mat_std = np.std(loss_dist_mat,axis=0)

    fig3, axs = plt.subplots(3,2, figsize=(15,15))
    plt.subplots_adjust(wspace=0.4, hspace=0.4)

    plt.rc('text', usetex=True)

    indx = 0
    axs[0,0].hist(loss_dist_mat[:,indx], int(loss_dist_mat.shape[0]/4), density = True, color = 'lightskyblue', edgecolor='black')
    axs[0,0].set_title(r'\textbf{Distribution of Rotation Error:} $\theta - \hat{\theta}$')
    axs[0,0].grid(False)
    axs[0,0].set_ylabel('PDF')
    axs[0,0].set_xlabel('Error in Angle Pred (radians)')
#     axs[0,0].text(-4.5, 0.31, 
#                   r'$\mu =\ $' + str(np.around(loss_dist_mat_mean[indx],3)) + '\n' + r'$\sigma =\ $' + str(np.around(loss_dist_mat_std[indx],3)))

    indx = 1
    axs[1,0].hist(loss_dist_mat[:,indx], int(loss_dist_mat.shape[0]/4), density = True, color = 'lightskyblue', edgecolor='black')
    axs[1,0].set_title(r'\textbf{Distribution of Rotation Error:} $\phi - \hat{\phi}$')
    axs[1,0].grid(False)
    axs[1,0].set_ylabel('PDF')
    axs[1,0].set_xlabel('Error in Angle Pred (radians)')
#     axs[1,0].text(-0.55, 2.6, 
#                   r'$\mu =\ $' + str(np.around(loss_dist_mat_mean[indx],3)) + '\n' + r'$\sigma =\ $' + str(np.around(loss_dist_mat_std[indx],3)))

    indx = 2
    axs[2,0].hist(loss_dist_mat[:,indx], int(loss_dist_mat.shape[0]/4), density = True, color = 'lightskyblue', edgecolor='black')
    axs[2,0].set_title(r'\textbf{Distribution of Rotation Error:} $\psi - \hat{\psi}$')
    axs[2,0].grid(False)
    axs[2,0].set_ylabel('PDF')
    axs[2,0].set_xlabel('Error in Angle Pred (radians)')
#     axs[2,0].text(-3.1, 1.1, 
#                   r'$\mu =\ $' + str(np.around(loss_dist_mat_mean[indx],3)) + '\n' + r'$\sigma =\ $' + str(np.around(loss_dist_mat_std[indx],3)))

    indx = 3
    axs[0,1].hist(loss_dist_mat[:,indx], int(loss_dist_mat.shape[0]/4), density = True, color = 'lightcoral', edgecolor='black')
    axs[0,1].set_title(r'\textbf{Distribution of Translation Error:} $x - \hat{x}$')
    axs[0,1].grid(False)
    axs[0,1].set_ylabel('PDF')
    axs[0,1].set_xlabel('Error in Trans Pred (mm)')
#     axs[0,1].text(-110, 0.013, 
#                   r'$\mu =\ $' + str(np.around(loss_dist_mat_mean[indx],3)) + '\n' + r'$\sigma =\ $' + str(np.around(loss_dist_mat_std[indx],3)))

    indx = 4
    axs[1,1].hist(loss_dist_mat[:,indx], int(loss_dist_mat.shape[0]/4), density = True, color = 'lightcoral', edgecolor='black')
    axs[1,1].set_title(r'\textbf{Distribution of Translation Error:} $y - \hat{y}$')
    axs[1,1].grid(False)
    axs[1,1].set_ylabel('PDF')
    axs[1,1].set_xlabel('Error in Trans Pred (mm)')
#     axs[1,1].text(-50, 0.018, 
#                   r'$\mu =\ $' + str(np.around(loss_dist_mat_mean[indx],3)) + '\n' + r'$\sigma =\ $' + str(np.around(loss_dist_mat_std[indx],3)))

    indx = 5
    axs[2,1].hist(loss_dist_mat[:,indx], int(loss_dist_mat.shape[0]/4), density = True, color = 'lightcoral', edgecolor='black')
    axs[2,1].set_title(r'\textbf{Distribution of Translation Error:} $z - \hat{z}$')
    axs[2,1].grid(False)
    axs[2,1].set_ylabel('PDF')
    axs[2,1].set_xlabel('Error in Trans Pred (mm)')
#     axs[2,1].text(-160, 0.009, 
#                   r'$\mu =\ $' + str(np.around(loss_dist_mat_mean[indx],3)) + '\n' + r'$\sigma =\ $' + str(np.around(loss_dist_mat_std[indx],3)))

    fig3.suptitle(r'\textbf{' + plot_title + '}',fontsize=16, y = 0.93, fontweight = 'bold')
    fig3.show()
    fig3.savefig(os.path.join(image_dir, plot_save_name), dpi=300)
    
    return None


```
